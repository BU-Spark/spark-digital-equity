```python
import pandas as pd
```

```python
course_enrollment = pd.read_csv('initial_data/CS-Course-Enrollment-2017.csv')
```

```python
course_enrollment
```

```python
course_enrollment = course_enrollment.drop(columns=["# STUDENTS ENROLLED PRIOR TO SECONDARY LEVEL", "# STUDENTS ENROLLED AT SECONDARY LEVEL"])
```

```python
course_enrollment
```

```python
demographic = pd.read_csv('initial_data/AP-CS-Gender-Ethnicity.xlsx-2017.csv')
```

```python
demographic
```

```python
demographic = demographic.drop(columns=['DIST_NAME', 'SCH_NAME'])
```

```python
demographic.dtypes
```

```python
scores = pd.read_csv('initial_data/2017-AP-Performance.csv')
```

```python
scores
```

```python
scores = scores.drop(columns=['School Name', '% Score 1-2', '% Score 3-5'])
```

```python
scores.dtypes
```

```python
all_merged = course_enrollment.merge(demographic, how='left', on='School Code').merge(scores, how='left', on='School Code')
```

```python
all_merged
```

```python
all_merged.dtypes
```

```python
all_merged = all_merged.astype({"STUDENTS_ENROLLED": "Int64", "ORG_CODE": "Int64", "AA": "Int64"})
```

```python
all_merged.dtypes
```

```python
##"AS": "Int64", "HI": "Int64", "MR": "Int64", "NA": "Int64", "NH_PI": "Int64", "WH": "Int64","EcoDis": "Int64","SWD": "Int64","ELL": "Int64","FEMALE": "Int64","MALE": "Int64", "NONBINARY": "Int64" 
```

```python
all_merged['SY'] = 2017
```

```python
all_merged
```

```python
all_merged.to_csv("all_merged.csv", index=False)
```

```python
all_merged_json = all_merged.to_json(orient='records')
```

```python
json_file = open("all_merged.json", "w")
json_file.write(all_merged_json)
json_file.close()
```
