```python
import pandas as pd
import numpy as np
```

```python
demographic = pd.read_csv('initial_data/APCS Gender & Ethnicity 2016.csv')
```

```python
demographic
```

```python
demographic = demographic.drop(columns=['# OF TEST_TAKERS', 'PCT'])
```

```python
demographic = demographic.rename(columns={"ECODIS": "EcoDis","SCHOOL": "School Code"})
```

```python
demographic
```

```python
demographic.dtypes
```

```python
demographic["STUDENTS_ENROLLED"] = pd.to_numeric(demographic["STUDENTS_ENROLLED"], errors='coerce')
demographic["MALE"] = pd.to_numeric(demographic["MALE"], errors='coerce')
demographic["WH"] = pd.to_numeric(demographic["WH"], errors='coerce')
```

```python
demographic["AA"] = demographic["AA"].astype("Int64")
demographic["AS"] = demographic["AS"].astype("Int64")
demographic["HI"] = demographic["HI"].astype("Int64")
demographic["MR"] = demographic["MR"].astype("Int64")
demographic["NA"] = demographic["NA"].astype("Int64")
demographic["NH_PI"] = demographic["NH_PI"].astype("Int64")
demographic["WH"] = demographic["WH"].astype("Int64")
demographic["EcoDis"] = demographic["EcoDis"].astype("Int64")
demographic["SWD"] = demographic["SWD"].astype("Int64")
demographic["ELL"] = demographic["ELL"].astype("Int64")

demographic["STUDENTS_ENROLLED"] = demographic["STUDENTS_ENROLLED"].astype("Int64")
demographic["MALE"] = demographic["MALE"].astype("Int64")
demographic["FEMALE"] = demographic["FEMALE"].astype("Int64")
```

```python
demographic.dtypes
```

```python
scores = pd.read_csv('initial_data/ap_performance-2016.csv')
```

```python
scores
```

```python
scores = scores.drop(columns=['School Name', '% Score 1-2', '% Score 3-5'])
```

```python
scores.dtypes
```

```python
all_merged = demographic.merge(scores, how='left', on='School Code')
```

```python
all_merged
```

```python
all_merged.dtypes
```

```python
all_merged = all_merged.astype({"STUDENTS_ENROLLED": "Int64", "ORG_CODE": "Int64"})
```

```python
all_merged["Tests Taken"] = all_merged["Tests Taken"].astype("Int64")
all_merged["Score=1"] = all_merged["Score=1"].astype("Int64")
all_merged["Score=2"] = all_merged["Score=2"].astype("Int64")
all_merged["Score=3"] = all_merged["Score=3"].astype("Int64")
all_merged["Score=4"] = all_merged["Score=4"].astype("Int64")
all_merged["Score=5"] = all_merged["Score=5"].astype("Int64")
```

```python
all_merged.dtypes
```

```python
all_merged['SY'] = 2016
```

```python
all_merged = all_merged.rename(columns={"DIST_NAME": "District","SCH_NAME": "School Name", "STUDENTS_ENROLLED": "# STUDENTS_ENROLLED" })
```

```python
all_merged
```

```python
all_merged
```

```python
all_merged.to_csv("all_merged.csv", index=False)
```

```python
all_merged_json = all_merged.to_json(orient='records')
```
